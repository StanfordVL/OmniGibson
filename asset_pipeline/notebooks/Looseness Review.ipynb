```python
# Same object, sometime loose, sometimes not
import sys
sys.path.append("/scr/BEHAVIOR-1K/asset_pipeline")
import b1k_pipeline.utils

for fn in glob.glob("/scr/BEHAVIOR-1K/asset_pipeline/cad/scenes/*/artifacts/object_list.json"):
    with open(fn) as f:
        scene_meshes = json.load(f)["meshes"]
    has_loose = set()
    has_fixed = set()
    for x in scene_meshes:
        m = b1k_pipeline.utils.parse_name(x)
        assert m, x
        add_to = has_loose if m.group("loose") else has_fixed
        key = f'{m.group("category")}-{m.group("model_id")}'
        add_to.add(key)
    conflicting = has_loose & has_fixed
    if conflicting:
        print(fn, conflicting)
```
